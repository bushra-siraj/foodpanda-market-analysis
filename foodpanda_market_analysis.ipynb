```python
top_5_restaurants = df['restaurant_name'].value_counts().head(5).index.tolist()
display(top_5_restaurants)
```

```python
inactive_customers = df[df['churned'] == 'Inactive']
inactive_restaurant_metrics = inactive_customers.groupby('restaurant_name').agg(
    Count_of_Inactive_Orders=('order_id', 'count'),
    Average_Price_per_Order_Inactive=('price', 'mean'),
    Average_Rating_Given_Inactive=('rating', 'mean')
).sort_values(by='Count_of_Inactive_Orders', ascending=False).head(5)

display(inactive_restaurant_metrics)
```

```python
average_orderfrequency_by_churn = df.groupby('churned')['order_frequency'].mean().reset_index()

display(average_orderfrequency_by_churn)
```

```python
category_metrics = df.groupby('category').agg(
    Total_Quantity_Sold=('quantity', 'sum'),
    Average_Price_per_Order=('price', 'mean')
).sort_values(by='Total_Quantity_Sold', ascending=False).head(3)

display(category_metrics)
```

```python
average_loyalty_by_age = df.groupby('age')['loyalty_points'].mean().reset_index()
display(average_loyalty_by_age)
```

```python
churn_by_age = df.groupby(['age', 'churned'], observed=False).size().unstack(fill_value=0)
churn_percentage_by_age = (churn_by_age.div(churn_by_age.sum(axis=1), axis=0)) * 100

plt.figure(figsize=(10, 6))
churn_percentage_by_age[['Active', 'Inactive']].plot(kind='bar', stacked=True, color=['#a63a50', '#B24C2C'], ax=plt.gca())
plt.title('Churn Percentage by Age Group', fontsize=16, fontweight='bold')
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Percentage of Customers', fontsize=12)
plt.xticks(rotation=0)
plt.yticks([0, 25, 50, 75, 100])
plt.legend(title='Churn Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(6, 7))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Customers by Gender', fontsize=16, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Order Prices', fontsize=16, fontweight='bold')
plt.xlabel('Price ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.countplot(data=df, y='category', order=df['category'].value_counts().index, hue='category', palette='crest')
plt.title('Distribution of Orders by Dish Category', fontsize=16, fontweight='bold')
plt.xlabel('Number of Orders', fontsize=12)
plt.ylabel('Dish Category', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
bad_order_counts = df[df['delivery_status'].isin(['Delayed', 'Cancelled'])].groupby('restaurant_name').size()

total_order_counts = df.groupby('restaurant_name').size()

bad_order_percentage = (bad_order_counts / total_order_counts) * 100

order_summary_simple = pd.DataFrame({
    'bad_orders': bad_order_counts,
    'total_orders': total_order_counts,
    'bad_order_percentage': bad_order_percentage
})

order_summary = order_summary_simple.sort_values(by='bad_order_percentage', ascending=False)
```

```python
top_5_restaurants = order_summary.sort_values(by='total_orders', ascending=False).head(5).index.tolist()
filtered_order_summary = order_summary[order_summary.index.isin(top_5_restaurants)]
```

```python
filtered_order_summary_sorted = filtered_order_summary.sort_values(by='bad_order_percentage', ascending=False)
display(filtered_order_summary_sorted)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=filtered_order_summary_sorted,
    x= bad_order_percentage,
    y= top_5_restaurants,
    palette='Spectral',
    hue='restaurant_name',
    legend=False)
plt.title('Top 5 restaurants by Bad order(Delayed/ Cancelled)', fontsize=14)
plt.xlabel('Bad Order Percentage (%)', fontsize=12)
plt.ylabel('Restaurant Name', fontsize=12)
plt.xlim(60, 70)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
city_order_counts = df['city'].value_counts().reset_index()
city_order_counts.columns = ['city', 'order_count']

plt.figure(figsize=(10, 6))
sns.barplot(x='city', y='order_count', data=city_order_counts, hue='city', palette='vlag')
plt.title('Number of Orders per City', fontsize=16, fontweight='bold')
plt.xlabel('City', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)
plt.xticks(rotation=0, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=average_orderfrequency_by_churn, x='churned', y='order_frequency', hue='churned', palette='coolwarm', legend=False)
plt.title('Average Order Frequency by Churn Status', fontsize=14, fontweight='bold')
plt.xlabel('Churn Status', fontsize=12)
plt.ylabel('Average Order Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
bins = [0, 167, 334, 500]
labels = ['Low', 'Medium', 'High']
df['loyalty_tier'] = pd.cut(df['loyalty_points'], bins=bins, labels=labels, right=True, include_lowest=True)

churn_by_loyalty_tier = df.groupby(['loyalty_tier', 'churned'], observed=False).size().unstack(fill_value=0)

churn_percentage_by_loyalty_tier = (churn_by_loyalty_tier.div(churn_by_loyalty_tier.sum(axis=1), axis=0)) * 100
display(churn_percentage_by_loyalty_tier)
```

```python
bins = [0, 167, 334, 500]
labels = ['Low', 'Medium', 'High']

if 'loyalty_tier' not in df.columns:
    df['loyalty_tier'] = pd.cut(df['loyalty_points'], bins=bins, labels=labels, right=True, include_lowest=True)

plt.figure(figsize=(10, 6))

sns.boxplot(
    data=df,
    x='loyalty_tier',
    y='loyalty_points',
    order=labels,
    hue='churned',
    palette='YlOrBr'
)

plt.title('Distribution of Loyalty Points by Tier', fontsize=14, fontweight='bold')
plt.xlabel('Loyalty Points Tier', fontsize=12)
plt.ylabel('Loyalty Points', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
orders_over_time = df.groupby(df['order_date'].dt.date).size().reset_index(name='order_count')
orders_over_time.columns = ['order_date', 'order_count']

plt.figure(figsize=(14, 7))
sns.lineplot(x='order_date', y='order_count', data=orders_over_time, color='teal')
plt.title('Number of Orders Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)
plt.xticks(rotation=0)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
df['bad_order'] = df['delivery_status'].apply(lambda x: 1 if x in ['Delayed', 'Cancelled'] else 0)

customer_summary = df.groupby('customer_id').agg(
    Total_Orders=('order_id', 'count'),
    Bad_Orders_Count=('bad_order', 'sum'),
    churned=('churned', 'first')
).reset_index()

customer_summary['Bad_Order_Percentage'] = (customer_summary['Bad_Orders_Count'] / customer_summary['Total_Orders']) * 100

median_bad_order= customer_summary['Bad_Order_Percentage'].median()

customer_summary['Service_Quality_Segment'] = np.where(
    customer_summary['Bad_Order_Percentage'] >= median_bad_order,
    'High Service Issue',
    'Low Service Issue'
)

churn_by_service = customer_summary.groupby(['Service_Quality_Segment', 'churned'], observed=False).size().unstack(fill_value=0)
churn_by_service = churn_by_service.div(churn_by_service.sum(axis=1), axis=0) * 100
churn_by_service = churn_by_service[['Active', 'Inactive']]
display(churn_by_service)
```

```python
churn_by_service.plot(
    kind='bar',
    stacked=True,
    color=['#D4C033', '#2B468A'],
    figsize=(9, 6)
)
plt.title('Customer Churn Rate by Service Quality Segment', fontsize=14, fontweight='bold')
plt.xlabel('Customer Service Quality Segment', fontsize=12)
plt.ylabel('Percentage of Customers (%)', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Churn Status', loc='upper right')
plt.yticks([0, 25, 50, 75, 100])
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.tight_layout()
plt.show()
```
